```python
# =================================================================
# PROGRAMA: SISTEMA IA - GESTIÓN DE CIENTÍFICOS (POO INTEGRADA)
# Estructura: Todas las funciones operativas dentro de la clase
# =================================================================

# Contenedor global de objetos
db_cientificos = {}

class Cientifico:
    """Clase que representa el molde y la gestión de los Científicos."""

    def __init__(clase, rama, exp, pat, equipo, presupuesto, tesis):
        # Atributos de instancia
        clase.rama_ciencia = rama
        clase.experimentos_activos = exp
        clase.patentes_registradas = pat
        clase.equipo_laboratorio = equipo  # Atributo tipo lista
        clase.presupuesto_investigacion = presupuesto
        clase.tesis_dirigidas = tesis

    # --- MÉTODOS DE OBTENCIÓN DE DATOS ---
    def obtener_equipo(clase):
        return " / ".join(clase.equipo_laboratorio)

    def obtener_presupuesto(clase):
        return f"${clase.presupuesto_investigacion}"

    # --- FUNCIONES DE GESTIÓN (DENTRO DE LA CLASE) ---

    def agregar_medico(clase):
        """Captura datos y genera un nuevo objeto en el diccionario."""
        print("\n--- Formulario de Entrada de Datos ---")
        rama = input("Rama de la ciencia: ")
        exp = input("Experimentos activos: ")
        pat = input("Número de patentes: ")

        print("Mencione 2 equipos de laboratorio:")
        e1 = input(" - Equipo 1: ")
        e2 = input(" - Equipo 2: ")
        lista_equipo = [e1, e2]

        pres = input("Presupuesto de investigación: ")
        tesis = input("Tesis dirigidas: ")

        # Creamos la instancia y la guardamos en el diccionario global
        nuevo_obj = Cientifico(rama, exp, pat, lista_equipo, pres, tesis)
        db_cientificos[rama] = nuevo_obj
        print(f"\n✅ Científico en {rama} guardado exitosamente.")

    def listar_medicos(clase):
        """Presenta los objetos registrados en una tabla alineada."""
        if not db_cientificos:
            print("\n[!] No hay científicos registrados en el sistema.")
        else:
            print("\n" + "━"*160)
            header = f"{'RAMA CIENCIA':<20} | {'EXP.':<8} | {'PATENTES':<10} | {'PRESUPUESTO':<15} | {'TESIS':<10} | {'EQUIPO DE LABORATORIO'}"
            print(header)
            print("━"*160)

            for key in db_cientificos:
                obj = db_cientificos[key]
                fila = (f"{obj.rama_ciencia:<20} | "
                        f"{obj.experimentos_activos:^8} | "
                        f"{obj.patentes_registradas:^10} | "
                        f"{obj.obtener_presupuesto():<15} | "
                        f"{obj.tesis_dirigidas:^10} | "
                        f"{obj.obtener_equipo()}")
                print(fila)
            print("━"*160)

    def actualizar_medico(clase):
        """Busca el objeto y modifica su presupuesto."""
        target = input("\nIngrese la Rama de ciencia a actualizar: ")
        if target in db_cientificos:
            nuevo_pres = input("Ingrese el nuevo presupuesto ($): ")
            db_cientificos[target].presupuesto_investigacion = nuevo_pres
            print(f"✅ Presupuesto actualizado para {target}.")
        else:
            print("[!] No se encontró el registro.")

    def borrar_medico(clase):
        """Elimina el objeto del diccionario global."""
        target = input("\nIngrese la Rama de ciencia a borrar: ")
        if target in db_cientificos:
            del db_cientificos[target]
            print(f"🗑️ El registro de {target} ha sido eliminado.")
        else:
            print("[!] Error: El registro no existe.")

# --- INICIALIZACIÓN DE LA INTERFAZ ---

# Creamos un objeto base "gestor" para acceder a las funciones de la clase
gestor = Cientifico("", "", "", [], "", "")

def mostrar_menu():
    print("\n" + "═"*100)
    print(" 🔬 SISTEMA DE INTELIGENCIA CIENTÍFICA - MÉTODOS DENTRO DE CLASE 🔬 ")
    print("═"*100)
    print(" 1. agregar_medico()    -> (Registrar)")
    print(" 2. listar_medicos()     -> (Ver Tabla)")
    print(" 3. actualizar_medico()  -> (Editar)")
    print(" 4. borrar_medico()      -> (Eliminar)")
    print(" 5. Salir")
    print("═"*100)

# --- CICLO PRINCIPAL ---
while True:
    mostrar_menu()
    opcion = input("Seleccione una función (1-5): ")

    if opcion == "1":
        gestor.agregar_medico()
    elif opcion == "2":
        gestor.listar_medicos()
    elif opcion == "3":
        gestor.actualizar_medico()
    elif opcion == "4":
        gestor.borrar_medico()
    elif opcion == "5":
        print("\nCerrando sistema de IA Científica... ¡Éxito investigadores!")
        break
    else:
        print("\n[!] Opción inválida. Intente de nuevo.")
```
