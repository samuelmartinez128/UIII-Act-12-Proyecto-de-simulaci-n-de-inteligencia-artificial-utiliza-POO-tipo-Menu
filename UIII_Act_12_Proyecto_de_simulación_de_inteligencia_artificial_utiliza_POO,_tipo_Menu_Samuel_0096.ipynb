```python
# =================================================================
# PROGRAMA: SISTEMA IA - GESTIÓN DE CIENTÍFICO (POO)
# Estructura: Clase Cientifico y Diccionario de Objetos
# =================================================================

class Cientifico:
    """Clase que representa el molde para cada Científico."""

    def __init__(self, rama, experimentos, patentes, equipo, presupuesto, tesis):
        # Atributos de la clase
        self.rama = rama
        self.experimentos = experimentos
        self.patentes = patentes
        self.equipo = equipo  # Recibe una lista
        self.presupuesto = presupuesto
        self.tesis = tesis

# --- CONTENEDOR PRINCIPAL ---
# El diccionario empieza vacío. Guardará objetos de la clase Cientifico.
db_cientificos = {}

def mostrar_menu():
    print("\n" + "═"*90)
    print(" 🔬 MENÚ DE INTELIGENCIA CIENTÍFICA (ORIENTADO A OBJETOS) 🔬 ")
    print("═"*90)
    print(" 1. Agregar Científico")
    print(" 2. Listar Científicos (Tabla)")
    print(" 3. Actualizar Datos (Presupuesto)")
    print(" 4. Borrar Científico")
    print(" 5. Salir")
    print("═"*90)

def agregar_medico():
    """Crea un objeto Cientifico y lo guarda en el diccionario."""
    print("\n--- Registro de Nuevo Científico ---")
    rama = input("Rama de la ciencia: ")
    exp = input("Experimentos activos: ")
    pat = input("Número de patentes: ")

    # Atributo lista
    print("Mencione 2 equipos clave:")
    eq1 = input(" - Equipo 1: ")
    eq2 = input(" - Equipo 2: ")
    lista_equipo = [eq1, eq2]

    pres = input("Presupuesto de investigación: ")
    tesis = input("Tesis dirigidas: ")

    # Instanciamos el objeto
    nuevo_cientifico = Cientifico(rama, exp, pat, lista_equipo, pres, tesis)

    # Guardamos en el diccionario usando la rama como Key
    db_cientificos[rama] = nuevo_cientifico
    print(f"\n✅ Objeto '{rama}' registrado exitosamente.")

def listar_medicos():
    """Recorre el diccionario de objetos y los muestra en tabla."""
    if not db_cientificos:
        print("\n[!] No hay científicos registrados.")
    else:
        print("\n" + "━"*150)
        header = f"{'RAMA CIENCIA':<20} | {'EXP.':<8} | {'PATENTES':<10} | {'PRESUPUESTO':<15} | {'TESIS':<10} | {'EQUIPO DE LABORATORIO'}"
        print(header)
        print("━"*150)

        for rama in db_cientificos:
            obj = db_cientificos[rama]
            equipos = " / ".join(obj.equipo)
            print(f"{obj.rama:<20} | {obj.experimentos:^8} | {obj.patentes:^10} | ${obj.presupuesto:<14} | {obj.tesis:^10} | {equipos}")
        print("━"*150)

def actualizar_medico():
    """Busca un objeto en el diccionario y modifica sus atributos."""
    nombre_rama = input("\nIngrese la Rama del científico a actualizar: ")
    if nombre_rama in db_cientificos:
        nuevo_pres = input("Ingrese el nuevo Presupuesto de investigación: ")
        db_cientificos[nombre_rama].presupuesto = nuevo_pres
        print(f"✅ Datos actualizados para el objeto {nombre_rama}.")
    else:
        print("[!] Registro no encontrado.")

def borrar_medico():
    """Elimina el objeto del diccionario."""
    nombre_rama = input("\nIngrese la Rama del científico a borrar: ")
    if nombre_rama in db_cientificos:
        del db_cientificos[nombre_rama]
        print(f"🗑️ Registro de {nombre_rama} eliminado del sistema.")
    else:
        print("[!] Error: No se encontró la clave.")

# --- CICLO DE CONTROL (While True) ---
while True:
    mostrar_menu()
    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregar_medico()
    elif opcion == "2":
        listar_medicos()
    elif opcion == "3":
        actualizar_medico()
    elif opcion == "4":
        borrar_medico()
    elif opcion == "5":
        print("\nSaliendo del sistema de IA. ¡Ciencia aplicada!")
        break
    else:
        print("\n[!] Opción inválida. Intente de nuevo.")
```
