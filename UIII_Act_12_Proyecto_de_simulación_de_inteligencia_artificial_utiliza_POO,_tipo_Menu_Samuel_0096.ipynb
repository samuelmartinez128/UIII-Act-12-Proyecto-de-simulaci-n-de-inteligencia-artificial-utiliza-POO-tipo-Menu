```python
# =================================================================
# PROGRAMA: SISTEMA IA - GESTIÓN DE CIENTÍFICOS (POO)
# Estructura: Clase Cientifico con métodos internos y "clase" como puntero
# =================================================================

class Cientifico:
    """Clase que representa el molde para cada Científico."""

    def __init__(clase, rama, exp, pat, equipo, presupuesto, tesis):
        # Atributos de la clase
        clase.rama_ciencia = rama
        clase.experimentos_activos = exp
        clase.patentes_registradas = pat
        clase.equipo_laboratorio = equipo  # Recibe una lista
        clase.presupuesto_investigacion = presupuesto
        clase.tesis_dirigidas = tesis

    # --- FUNCIONES (MÉTODOS) DENTRO DE LA CLASE ---
    # Estas funciones retornan o procesan los atributos del objeto

    def obtener_rama(clase):
        return clase.rama_ciencia

    def obtener_exp(clase):
        return clase.experimentos_activos

    def obtener_patentes(clase):
        return clase.patentes_registradas

    def obtener_equipo(clase):
        return " / ".join(clase.equipo_laboratorio)

    def obtener_presupuesto(clase):
        return f"${clase.presupuesto_investigacion}"

    def obtener_tesis(clase):
        return clase.tesis_dirigidas

# --- CONTENEDOR PRINCIPAL ---
# Diccionario vacío que almacenará los objetos instanciados
db_cientificos = {}

def mostrar_menu():
    print("\n" + "═"*100)
    print(" 🔬 SISTEMA DE INTELIGENCIA CIENTÍFICA - GESTIÓN DE OBJETOS 🔬 ")
    print("═"*100)
    print(" 1. agregar_medico()    -> (Registrar Científico)")
    print(" 2. listar_medicos()     -> (Ver Bitácora en Tabla)")
    print(" 3. actualizar_medico()  -> (Modificar Presupuesto)")
    print(" 4. borrar_medico()      -> (Eliminar Registro)")
    print(" 5. Salir")
    print("═"*100)

def agregar_medico():
    """Crea un objeto Cientifico y lo guarda en el diccionario."""
    print("\n--- Formulario de Entrada de Datos ---")
    rama = input("Rama de la ciencia: ")
    exp = input("Experimentos activos: ")
    pat = input("Número de patentes: ")

    # Captura de lista para el equipo de laboratorio
    print("Mencione 2 equipos de laboratorio:")
    e1 = input(" - Equipo 1: ")
    e2 = input(" - Equipo 2: ")
    lista_equipo = [e1, e2]

    pres = input("Presupuesto de investigación: ")
    tesis = input("Tesis dirigidas: ")

    # INSTANCIACIÓN: Creamos el objeto usando la clase Cientifico
    nuevo_obj = Cientifico(rama, exp, pat, lista_equipo, pres, tesis)

    # Guardamos en el diccionario usando la 'rama' como Key
    db_cientificos[rama] = nuevo_obj
    print(f"\n✅ Científico en {rama} guardado como objeto en el diccionario.")

def listar_medicos():
    """Presenta la información de los objetos en una tabla alineada."""
    if not db_cientificos:
        print("\n[!] No hay científicos registrados en el sistema.")
    else:
        print("\n" + "━"*160)
        # Encabezados de la tabla
        header = f"{'RAMA CIENCIA':<20} | {'EXP.':<8} | {'PATENTES':<10} | {'PRESUPUESTO':<15} | {'TESIS':<10} | {'EQUIPO DE LABORATORIO'}"
        print(header)
        print("━"*160)

        # Recorremos el diccionario para extraer cada objeto
        for key in db_cientificos:
            cientifico_obj = db_cientificos[key]

            # Llamamos a las funciones DENTRO de la clase para obtener los valores
            fila = (f"{cientifico_obj.obtener_rama():<20} | "
                    f"{cientifico_obj.obtener_exp():^8} | "
                    f"{cientifico_obj.obtener_patentes():^10} | "
                    f"{cientifico_obj.obtener_presupuesto():<15} | "
                    f"{cientifico_obj.obtener_tesis():^10} | "
                    f"{cientifico_obj.obtener_equipo()}")
            print(fila)
        print("━"*160)

def actualizar_medico():
    """Busca el objeto por su Key y actualiza el presupuesto."""
    target = input("\nIngrese la Rama de ciencia del registro a actualizar: ")
    if target in db_cientificos:
        nuevo_pres = input("Ingrese el nuevo presupuesto de investigación ($): ")
        # Actualizamos el atributo directamente en el objeto
        db_cientificos[target].presupuesto_investigacion = nuevo_pres
        print(f"✅ Presupuesto actualizado para {target}.")
    else:
        print("[!] No se encontró ningún científico en esa rama.")

def borrar_medico():
    """Elimina la entrada del diccionario y destruye el objeto de la memoria."""
    target = input("\nIngrese la Rama de ciencia del registro a borrar: ")
    if target in db_cientificos:
        del db_cientificos[target]
        print(f"🗑️ El registro de {target} ha sido eliminado satisfactoriamente.")
    else:
        print("[!] Error: El registro no existe.")

# --- CICLO PRINCIPAL DE LA IA (While True) ---
while True:
    mostrar_menu()
    opcion = input("Seleccione una función (1-5): ")

    if opcion == "1":
        agregar_medico()
    elif opcion == "2":
        listar_medicos()
    elif opcion == "3":
        actualizar_medico()
    elif opcion == "4":
        borrar_medico()
    elif opcion == "5":
        print("\nCerrando sistema de IA Científica... ¡Éxito en sus experimentos!")
        break
    else:
        print("\n[!] Opción inválida. Intente con los números del menú.")
```
